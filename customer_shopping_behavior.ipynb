Now, to replace NA with rating values, we use Median instead of Mean, since Mean can be biased by outliers. Hence, when the data is skewed, we use the median.
Here we replace the ratings category wise so that one category  rating median does not go into another

```python
grouped = df.groupby('Category')
print(list(grouped))
#trial&error
```

```python
print(grouped.groups.keys())
#trial&error
```

```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
#snake casing column names
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#feature engineering , creating new columns
labels=['young-adult', 'adult', 'middle-aged' , 'Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head(11)
```

```python
#create new column purchase frequency days
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
```
